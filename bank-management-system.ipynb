```python
import random

customers = {}
balance = {}
account_numbers = set()
transactions={}
def generate_unique_account_number():
    while True:
        number = random.randint(10**11, 10**12 - 1)
        if number not in account_numbers:
            account_numbers.add(number)
            return number

def view_transactions():
    acc_no = int(input("Enter your account number: "))
    
    if acc_no not in customers:
        print("Account not found.")
        return

    print("=== Transaction History ===")
    user_transactions = transactions.get(acc_no, [])

    if not user_transactions:
        print("No transactions yet.")
    else:
        for i, t in enumerate(user_transactions, 1):
            print(f"{i}. {t}")

    print("============================\n")
def create_account():
    name = input("Enter your full name: ").title()
    age = int(input("Enter your age: "))
    gender = input("Enter your gender (Male/Female): ")
    phone_no = input("Enter your 10-digit phone number: ")

    acc_no = generate_unique_account_number()

    customers[acc_no] = {
        'name': name,
        'age': age,
        'gender': gender,
        'phone_no': phone_no,
    }
    balance[acc_no] = 0

    print(f"\n Account created successfully! Your account number is: {acc_no}\n")


def deposit():
    acc_no = int(input("Enter your account number: "))
    if acc_no in customers:
        amount = int(input("Enter deposit amount: "))
        if amount > 0:
            balance[acc_no] += amount
            transactions.setdefault(acc_no, []).append(f"Deposited : ₹{amount}")
            print(f" Deposited ₹{amount}. New Balance: ₹{balance[acc_no]}")
        else:
            print(" Invalid amount.")
    else:
        print(" Account not found.")


def withdraw():
    acc_no = int(input("Enter your account number: "))
    if acc_no in customers:
        amount = int(input("Enter withdrawal amount: "))
        if amount <= 0:
            print(" Enter a valid amount.")
        elif amount > balance[acc_no]:
            print(" Insufficient balance.")
        else:
            balance[acc_no] -= amount
            transactions.setdefault(acc_no, []).append(f"Withdrawn : ₹{amount}")
            print(f" Withdrawn ₹{amount}. New Balance: ₹{balance[acc_no]}")
    else:
        print(" Account not found.")


def view_balance():
    acc_no = int(input("Enter your account number: "))
    if acc_no in customers:
        print(f" Current Balance: ₹{balance[acc_no]}")
    else:
        print(" Account not found.")


def transfer():
    sender = int(input("Enter your account number: "))
    if sender not in customers:
        print(" Sender account not found.")
        return

    receiver = int(input("Enter the 12-digit recipient account number: "))
    if receiver not in customers:
        print(" Recipient account not found.")
        return

    print(f"Recipient Name: {customers[receiver]['name']}")
    confirm = input("Is the name correct? (yes/no): ").lower()

    if confirm != 'yes':
        print(" Transfer cancelled.")
        return

    amount = int(input("Enter transfer amount: "))
    if amount <= 0:
        print(" Invalid amount.")
    elif amount > balance[sender]:
        print(" Insufficient balance.")
    else:
        balance[sender] -= amount
        balance[receiver] += amount
        print(f" ₹{amount} transferred successfully.")


def view_account_details():
    acc_no = int(input("Enter your account number: "))
    if acc_no in customers:
        data = customers[acc_no]
        print(f"\n--- Account Details ---")
        print(f"Name     : {data['name']}")
        print(f"Age      : {data['age']}")
        print(f"Gender   : {data['gender']}")
        print(f"Phone No.: {data['phone_no']}")
        print(f"Balance  : ₹{balance[acc_no]}")
        print("------------------------\n")
    else:
        print(" Account not found.")


while True:
    print("===== Bank Management System ====")
    print("1. Create account")
    print("2. Deposit")
    print("3. Withdrawal")
    print("4. View balance")
    print("5. Transfer money")
    print("6. View account details")
    print("7. View transaction history")
    print("8. Exit")


    try:
        choice = int(input("Enter your choice (1-7): "))
    except ValueError:
        print(" Please enter a valid number.\n")
        continue

    if choice == 1:
        create_account()
    elif choice == 2:
        deposit()
    elif choice == 3:
        withdraw()
    elif choice == 4:
        view_balance()
    elif choice == 5:
        transfer()
    elif choice == 6:
        view_account_details()
    elif choice == 7:
        view_transactions()
    elif choice == 8:
        print(" Exiting the program. Thank you!")
        break

    else:
        print(" Invalid choice. Please try again.\n")
```
